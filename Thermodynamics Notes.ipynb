__ Absolute Pressure -__ Denoted (abs) pressure relative to absolute zero pressure (perfect vaccuum)

__ Bulk Modulus (of Elasticity) ($E_v$)-__ Measurement of compressibility. $E_v = -\frac{dp}{\frac{d\veebar}{\veebar}} = \frac{dp}{\frac{d\rho}{\rho}} $ dimensions of pressure $\frac{[F]}{[L]^2}$

__ Cavitation -__ The formation and subsequent collapse of vapor bubbles in a flowing fluid. In flowing fluids it is possible to develop very low pressure due to fluid motion. If the pressure drops to or below the __vapor pressure__ boiling will occur. If the vapor bubbles thus formed swept into a region of higher pressure they can suddenly collapse. This can happen with sufficent intensity to cause structural damage to a pump or or other such fluid system.

__ Continuum -__ Modeling an object as a continuum assumes that the substance *has average properties though the volume which it fills. These properties include density, specific weight, specific gravity etc. Liquids and gases are both treated as continuous fluids.

__ Compressed Liquid -__ Aka *subcooled liquid*, State 1: A liquid which is not about to vaporize. (it's below it's boiling point). 
* *Saturated liquid:* State 2: When a liquid reaches it's boiling point, but has not yet started to vaporize.
* Saturated liquid-vapor mixture:* state 3: midway between the saturated liquid state and a fully vapor state. Both saturated states co-exist.
* *Saturated Vapor:* State 4: Once the liquid is fully vaporized, any heat loss will cause it to condense, this is a saturated vapor.
* *Superheated Vapor:* State 5: If the vapor is above the condensation point and can lose heat without condensing.

__ Control Boundary -__ The real or imaginary boundaries of a control volume. 

__ Control Mass -__ Aka *closed system*: consits of a fixed amount of mass (no mass can cross system boundary). Energy in the form of heat and work can cross system boundary. If energy can not cross the boundary it is a special case called an isolated system.

__ Control Volume -__ Aka *open system*: A system where both mass and energy can cross the system boundary. The volume of the system can be fixed, moving, real, or imaginary. Useful for mass flow problems. 

__ Density -__ Mass per unit volume ${[\rho]}\doteq \frac{{[M]}}{[L]^3}$ For a differential volume element of mass: ${[\rho]}\doteq \frac{{[\delta m]}}{[\delta \veebar]}$ For most gasses, density is proportional to Pressure and inverely proportional to temp.

__ Dynamic Viscosity ($\mu$) -__ Constant of porportionality which relates __shear stess ($\tau$)__ with the __velocity gradiant ($\frac{du}{dy}$)__. dimensions are given in $\frac{[F][t]}{[L]^2}$

__ Fluid -__ A substance that __deforms continuously__ when acted on by a __shearing stress__ of any magnitude. Properties of a fluid include:
  * Needs a container
  * Takes the shape of the container
  * Sustains compression and tension
  * Cohesive forces cause fluid molecules to form a surface which is always horizontal
  
__ Gage Pressure -__ Pressure measured relative to local atmospheric pressure. Absolute pressure can be obtained by adding the atmospheric pressure to the gage pressure.

__ Isentropic -__ A process in which there is no heat exchanged with the surroundings. For gas expansion the relationship between pressure and density is $\frac{p}{\rho ^k}=$ a constant, where $k $ is the rato of the specific heat at a constant pressure ($C_p$) to the heat at a constant volume ($C_\veebar$) so $k=\frac{C_p}{C_\veebar} $

__ Iso- __ Prefix often used to designate a process for which a particular property remains constant.
* *Isothermal:* A process under constant temperature conditions. For gas expansion the relationship between pressure and density is $\frac{p}{\rho}=$ a constant.
* *Isobaric:* A process under constant pressure conditions.
* *Isochoric (or isometric):* A process under which the specific volume remains constant.
* *Isotropic:* A process with constant entrophy ($S$)
* *Isothalpic:* A process with constant enthalpy ($h$)
 
__ Kinetic Viscosity ($\nu$)-__ $\frac{\mu}{\rho}$ the ration between the __Dynamic Viscosity__ and the __Density__ of a fluid. This is the governing portion of the __Reynolds number__ dimensions $\frac{{[L]^2}}{[t]}$

__ Newtonian Fluid -__ Fluids for which the shearing stress is linearly related to the rate of shearing strain (aka the rate of angular deformation $\frac{\delta \beta}{\delta t}$).
 - _Shear thinning fluids:_ apparent viscosity decreases with increasing shear rate. More shear $\rightarrow$ less viscosity (easier flow)
 - _Shear thickening fluids:_ apparent viscosity increses with increasing shear rate. More shear $\rightarrow$ more viscosity (harder flow)
 
__Mach Number ($Ma$) -__ The ratio between velocity and the speed of sound in a liquid. if $Ma < 1.0$ speed is subsonic, if $Ma > 1.0$ speed is supersonic. This is a dimensionless parameter. Speed of sound increases with increasing temperature $\therefore$ Mach number decreases as temp increases. 

__ Path -__ The series of states that a system passes through during a process.

__ Pressure -__ Normal force per unit area ${[p]}\doteq \frac{{[F]}}{[L]^2}$
* *Pressure Head:* the change in pressure resulting from a change in depth or height of a fluid 

__ Process -__ The change of a system from one equilibrium state to another. To fully describe a process one needs the initial, final conditions, the __path__ and any interactions with the surroundings.

__ Properties -__ Any characteristic of a system:
 * *Intensive Properties:* Properties which do not vary with the mass of a system (temp, pressure, density)
 * *Extensive Properties:* Properties which depend on the size or extent of a system (mass, volume)
 * *Specific Properties:* Extensive properties when give as a per unit mass. i.e. specific volume ${[\upsilon]}\doteq \frac{{[L]^3}}{[M]} $

__ Quasi-equilibrium process -__ A process which takes place sufficently slow to allow the system to adjust itself internally so that properties in one part fo the system do not change faster than those at other parts.

__ Reynolds Number ($Re $)__ $\frac{\rho VD}{\mu}$ A dimensionless number used in fluid mechanics to indicate whether fluid flow past a body or in a duct is steady or turbulent.  Where V is mean fluid velocity, D is diameter of pipe. The ratio $\frac{\mu}{\rho}$ is the important piece of the Reynolds number See __Kinematic Viscosity__. 

__ Saturation Pressure -__ For a given temp, the pressure at which a pure substance changes phase.

__ Saturation Temperature -__ For a given pressure, the temp at which a pure substance changes phase.

__ Shearing Stress ($\tau$) -__ Tangential force acting on a surface, units given in force per unit area $\frac{[F]}{[A]^2}$ 

__ Speed of Sound -__ The velocity at which small disturbances propogate (pressure waves) through a fluid. Related to changes in pressure and density of fluid medium through eq: $c = \sqrt{\frac{dp}{d\rho}}$ and in terms of bulk modulus by $c = \sqrt{\frac{E_v}{d\rho}}$ *($\leftarrow$ can be used with liquids)*. If disturbance is small, heat transfer is negligible (__isentropic__). $\therefore c = \sqrt{\frac{kp}{\rho}}$ Using ideal gas law $c = \sqrt{kRT}$  

__ Specific Gravity ($SG$)-__ The ratio of the density of a fluid to the density of water at a specified temperature. ${[SG]}\doteq \frac{{[\rho]}}{[\rho]_{H2O}}$ 

__ Specific Volume ($\upsilon$) -__ Volume per unit mass, the reciprocal of __Density__ $\therefore$ ${[\upsilon]}\doteq \frac{{[L]^3}}{[M]}$

__ Specific Weight ($\gamma$) -__ Weight per unit volume, related to __Density__ through ${[\gamma]}\doteq [\rho] [g]$

__ State -__ a condition where all properties of a system have fixed values which can be measured or calculated.
* *Equilibrium State:* A state where there are no unbalanced potentials within the system, it experiences no changes when it is isolated from it's surroundings.
    - *Thermal Equilibrium:* When the temperature is the same throughout an entire system
    - *Mechanical Equilibrium:* Related to pressure, a system is in mechanical equilibrium if no change in pressure at any point of the system *with time*. The pressure can be different at different points within the system.
    - *Phase Equilibrium:* In a multiphase system, when the mass of each phase reaches an equilibrium level and stays there.
    - *Chemical Equilibrium:* When the chemical composition does not change with time.
    
__ State Postulate -__ "The state of a simple compressible system is completely specified by two independent, intensive properties" - In other words once a sufficient number of prooperties are specified, the rest of the properties assume certain values automatically and fixes the state. The properties must be independant. (one property can be varied while the other can be held constant.)
* *Simple Compressible System:* A system without electrical, magnetic, gravitational, motion, and surface tension effects. - If present, an additional property needs to be specified for each in order to fix the state. i.e. if gravity is to be considered, the height $h$ needs to be specified in addition to those from the __state postulate__.

__ Steady -__ no change with time. Oposite is unsteady aka __transient__. 
* *Steady-flow process:* a process during which a fluid flows through a control volume steadily. 

__ Surface Tension ($\sigma$-__ The intensity of the molecular attraction per unit length along any line in the free surface of a fluid. dimensions $\frac{[F]}{[L]}$

__Transient Motion -__ An oscillatory or other irregular motion occurring while a quantity is changing to a new steady-state value.

__ Uniform -__ No change with location over a specified region.

__ Vapor Pressure ($p_v$) -__  The pressure exerted by a vapor in thermodynamic equilibrium with its condensed phases (solid or liquid) at a given temperature in a closed system. The equilibrium vapor pressure is an indication of a liquid's evaporation rate. *A liquid boils when the absolute pressure becomes equal to the vapor pressure* 

__ Velosity Gradiant ($\frac{du}{dy}$) -__ A relationship between the speed of flow relative to distance from a surface in no slip conditions. In more complex situations this relationship is not linear.

__ Viscosity -__ See __Dynamic Viscosity__ and __Kinetic Viscosity__

### Equations

__*Liquids*__ 

__ Andrade's equation -__ $\mu = De^\frac{B}{T}$ - Relates temperature to viscosity. D and B are empirical constants, T is absolute Temperature. Viscosity must be known for two temperatures to solve for B & T.


__*Gasses*__

__ Sutherland equation -__ $\mu = \frac{CT^\frac{3}{2}}{T+S}$ - Relates Temperature to Viscosity. C and S are empirical constants, T is absolute Temp. If viscosity is known at two temperatures, C & S can be determined.
